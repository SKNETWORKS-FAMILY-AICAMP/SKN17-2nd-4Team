### 결측치 제거

```python
# 2. 각 pid 그룹 내에서 다음 연도의 label 값을 가져옵니다.
# df_all['next_year_label'] = df_all.groupby('pid')['label'].shift(-1)
# df_all
# label 결측치 제거
mask = df_all['label'].isna() | (df_all['label'].astype(str).str.strip() == '')
df_all = df_all.loc[~mask]
```

```python
# label의 다음 연도 값
df_all['next_year_label'] = df_all.groupby('pid')['label'].shift(-1)
df_all
```

```python
# next_year_label 결측치 제거
mask = df_all['next_year_label'].isna() | (df_all['next_year_label'].astype(str).str.strip() == '')
df_all = df_all.loc[~mask]
```

```python
print(df_all[['label', 'next_year_label']].dtypes)
print(df_all['label'].unique()[:10])
print(df_all['next_year_label'].unique()[:10])
```

```python
df_all['label'] = pd.to_numeric(df_all['label'], errors='coerce').astype(int)
df_all['next_year_label'] = pd.to_numeric(df_all['next_year_label'], errors='coerce').astype(int)
```

```python
# 3. 이탈(churn) 여부를 정의합니다.
# 조건 1: 현재 레이블이 2이고 (df_all['label'] == 2)
# 조건 2: 다음 해의 레이블이 존재하며 2가 아닌 경우 (df_all['next_year_label'].notna() & (df_all['next_year_label'] != 2)) 
# label = 1 이면 페이스북 유저
# 위 두 조건을 만족하면 이탈(1), 아니면 비이탈(0)로 정의합니다.
churn_conditions = [
    (df_all['label'] == 1) & (df_all['next_year_label'].notna()) & (df_all['next_year_label'] != 1),
    (df_all['label'] == 1) & (df_all['next_year_label'].notna()) & (df_all['next_year_label'] == 1)
]
churn_choices = [1, 0] # 1: 이탈 (Churn), 0: 비이탈 (Not Churn)

df_all['churn'] = np.select(churn_conditions, churn_choices, default=np.nan)

print(df_all.head())
print(df_all['churn'].value_counts(dropna=False))
```

```python
# 4. 분석에 사용할 데이터셋을 생성합니다.
# 'churn' 값이 NaN이 아닌 경우, 즉 현재 label이 2인 고객들의 데이터만 필터링합니다.
# 이 데이터셋이 EDA와 모델링의 기반이 됩니다.
df_model_base = df_all[df_all['churn'].notna()].copy()

# # 불필요해진 'next_year_label' 컬럼 삭제
# df_model_base.drop(columns=['next_year_label'], inplace=True)

# df_model_base['churn'] = df_model_base['churn'].astype(int)

# df_model_base = df_final[df_final['churn'].notna()].copy()

# 불필요해진 'next_year_label' 컬럼 삭제
# df_model_base.drop(columns=['next_year_label'], inplace=True)
# df_model_base['churn'] = df_model_base['churn'].astype(int)

df_model_base.drop(columns=['next_year_label'], inplace=True, errors='ignore')
df_model_base['churn'] = df_model_base['churn'].astype(int)
print(df_model_base.isnull().sum())
```

```python
mapping = {
    '1': 11,  # 하루에도 여러 번 → 빈도 가장 높음
    '2': 10,  # 하루 1번(매일)
    '3': 9,   # 하루 5회 이상
    '4': 8,  # 하루 3~4회
    '5': 7,  # 하루 1~2회
    '6': 6,   # 1주일에 5~6회
    '7': 5,   # 1주일에 3~4회
    '8': 4,   # 1주일에 1~2회 정도
    '9': 3,   # 월 1~3회
    '10': 2,   # 월 1회 미만
    '11': 1    # 분기/연간 1회 정도 → 빈도 가장 낮음
}

df_model_base['d01'] = df_model_base['d01'].map(mapping).fillna(0).astype(int)
df_model_base['d02'] = df_model_base['d02'].map(mapping).fillna(0).astype(int)
df_model_base['d03'] = df_model_base['d03'].map(mapping).fillna(0).astype(int)
df_model_base['d04'] = df_model_base['d04'].map(mapping).fillna(0).astype(int)

df_model_base['relig'] = df_model_base['relig'].apply(lambda x: 0 if str(x).strip() == '' else x)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_curve
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore', category=UserWarning)

# ===== 1) 데이터 =====
X = df_final.drop(['churn','pid'], axis=1)
y = df_final['churn'].astype(int)

# (핵심) object/category → 원-핫 인코딩 (트리에 스케일 필요 없음)
obj_cols = X.select_dtypes(include=['object','category']).columns.tolist()
if len(obj_cols):
    X = pd.get_dummies(X, columns=obj_cols, drop_first=True)

# 0-분산 컬럼 제거(모델 분할 방해)
zero_var = X.columns[X.nunique(dropna=False) <= 1]
if len(zero_var):
    X = X.drop(columns=zero_var)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42
)

print(f"[X shape] train={X_train.shape}, test={X_test.shape}")
print(f"[zero-var removed] {list(zero_var)}")

print("=" * 20, "Random Forest 모델", "=" * 20)

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=4,
    min_samples_leaf=10,   # 30 → 12 (데이터 75개 기준 너무 컸음)
    min_samples_split=24,  # 300 → 24 (분할 가능하게 낮춤)
    max_features='sqrt',
    class_weight='balanced',
    oob_score=True,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

# ===== 3) 평가 =====
proba_tr = rf.predict_proba(X_train)[:, 1]
proba_te = rf.predict_proba(X_test)[:, 1]
pred_tr  = (proba_tr >= 0.5).astype(int)
pred_te  = (proba_te >= 0.5).astype(int)

print("===== Random Forest - Train Set Evaluation =====")
print(classification_report(y_train, pred_tr, digits=3))
print(f"ROC AUC Score: {roc_auc_score(y_train, proba_tr):.4f}")

print("\n===== Random Forest - Test Set Evaluation (thr=0.5) =====")
print(classification_report(y_test, pred_te, digits=3))
print(f"ROC AUC Score: {roc_auc_score(y_test, proba_te):.4f}")

# ===== 4) 확률 진단(왜 0.5만 나오는지 확인) =====
print("\n[proba stats] train min/max/unique_cnt:",
      np.min(proba_tr), np.max(proba_tr), len(np.unique(proba_tr)))
print("[proba stats] test  min/max/unique_cnt:",
      np.min(proba_te), np.max(proba_te), len(np.unique(proba_te)))

# ===== 5) 임계값 튜닝 (원하던 0.7 이상 기준 체크용) =====
from sklearn.metrics import f1_score
p, r, thr = precision_recall_curve(y_test, proba_te)
f1 = 2*p*r/(p+r+1e-12)
best_thr = thr[np.argmax(f1)] if len(thr) else 0.5
pred_te_best = (proba_te >= best_thr).astype(int)
print(f"\n===== Threshold Tuning =====\nBest thr={best_thr:.3f}")
print(classification_report(y_test, pred_te_best, digits=3))
```

```python
import numpy as np
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, roc_auc_score, roc_curve

# ===== 1) 파이프라인 (CV 안에서 SMOTE 적용) =====
pipe = Pipeline(steps=[
    ("smote", SMOTE(random_state=42)),
    ("clf", DecisionTreeClassifier(random_state=42))
])

param_grid = {
    "clf__criterion": ["gini", "entropy", "log_loss"],  # log_loss는 확률 보정에 유리한 경우 있음
    "clf__max_depth": [6, 10, 14, None],
    "clf__min_samples_split": [5, 10, 20, 40],
    "clf__min_samples_leaf": [3, 5, 10, 20],
    "clf__max_features": [None, "sqrt", "log2"],
    # 가지치기(비용-복잡도) 계수도 함께 탐색
    "clf__ccp_alpha": [0.0, 1e-5, 5e-5, 1e-4, 5e-4, 1e-3]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

gs = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring="roc_auc",      # 불균형에 강한 AUC 기준
    cv=cv,
    n_jobs=-1,
    verbose=0
)

gs.fit(X_train, y_train)
best_model = gs.best_estimator_
print("Best AUC (CV):", gs.best_score_)
print("Best Params:", gs.best_params_)

# ===== 2) 임계값 최적화 (Youden's J: TPR - FPR 최대) =====
proba_tr = best_model.predict_proba(X_train)[:, 1]
fpr, tpr, thresholds = roc_curve(y_train, proba_tr)
youden_j = tpr - fpr
best_thr = thresholds[np.argmax(youden_j)]
print(f"Best threshold (Youden J): {best_thr:.3f}")

# ===== 3) 최종 평가 (Train/Test) =====
proba_tr = best_model.predict_proba(X_train)[:, 1]
proba_te = best_model.predict_proba(X_test)[:, 1]
pred_tr  = (proba_tr >= best_thr).astype(int)
pred_te  = (proba_te >= best_thr).astype(int)

print("\n===== Train Set (DecisionTree + SMOTE-CV + Pruning) =====")
print(classification_report(y_train, pred_tr, digits=3))
print(f"ROC AUC: {roc_auc_score(y_train, proba_tr):.4f}")

print("\n===== Test Set (threshold tuned) =====")
print(classification_report(y_test, pred_te, digits=3))
print(f"ROC AUC: {roc_auc_score(y_test, proba_te):.4f}")
```

```python
from lightgbm import LGBMClassifier, early_stopping
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore', category=UserWarning)

# ===== 데이터 준비 =====
X = df_model_base.drop(['churn'], axis=1)
y = df_model_base['churn'].astype(int)

obj_cols = X.select_dtypes(include=['object']).columns
if len(obj_cols):
    X = pd.get_dummies(X, columns=obj_cols, drop_first=True)

zero_var = X.columns[X.nunique(dropna=False) <= 1]
if len(zero_var):
    X = X.drop(columns=zero_var)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42
)

# ===== SMOTE =====
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# ===== LightGBM 모델 =====
lgbm = LGBMClassifier(
    n_estimators=500,
    learning_rate=0.07,
    max_depth=5,
    num_leaves=20,
    min_child_samples=50,
    reg_alpha=0.1,
    reg_lambda=0.3,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

# ===== 학습 (callback 방식 early stopping) =====
lgbm.fit(
    X_train_res, y_train_res,
    eval_set=[(X_test, y_test)],
    eval_metric='auc',
    callbacks=[early_stopping(stopping_rounds=50)]
)

# ===== 평가 =====
proba_tr = lgbm.predict_proba(X_train)[:, 1]
proba_te = lgbm.predict_proba(X_test)[:, 1]
pred_tr  = (proba_tr >= 0.5).astype(int)
pred_te  = (proba_te >= 0.5).astype(int)

print("===== Train Set Evaluation =====")
print(classification_report(y_train, pred_tr, digits=3))
print(f"ROC AUC Score: {roc_auc_score(y_train, proba_tr):.4f}")

print("\n===== Test Set Evaluation (thr=0.5) =====")
print(classification_report(y_test, pred_te, digits=3))
print(f"ROC AUC Score: {roc_auc_score(y_test, proba_te):.4f}")
```
